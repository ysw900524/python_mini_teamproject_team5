### 데이터 수집 및 확인

```python
title = []
date = []
news = []
page = 0

start_flag = False
end_flag = False

end_date = input("시작 날짜를 입력하세요.  ex) 20181007 \n")
start_date = input("마지막 날짜를 입력하세요.  ex) 20181115 \n")
choose = input("제목만 가져오려면 : 1\n본문까지 가져오려면 : 2 \n")
end_date = end_date[:4] + '-' + end_date[4:6] + '-' + end_date[6:]
start_date = start_date[:4] + '-' + start_date[4:6] + '-' + start_date[6:]

print("=" * 50)

if choose == '1':
    print("[%s ~ %s] 뉴스(제목) 수집을 시작합니다.\n" % (end_date, start_date))
else:
    print("[%s ~ %s] 뉴스(제목+내용) 수집을 시작합니다.\n" % (end_date, start_date))

start = datetime.now()

for url in urls:
    html = urlopen(url)
    data = BeautifulSoup(html, "lxml")    
    data_temp = data.find_all('div', 'section-list-area')
    
    # 기사 제목 및 날짜 찾기
    data_title = data_temp[0].find_all('h4', 'article-title')
    data_date = data_temp[0].find_all('span', 'date')
#     data_date = data_date[0].get_text().split(' ')[0]
#     print(data_date)
    
    # 데이터수집 체크
    page += 1
    
    if start_flag:
        print("%d 페이지 수집중입니다.... ing" %page, end='')
    
    for x in range(len(data_title)):
        
        # 기사 끝날 날짜 체크 (end_flag 변경)
        if data_date[x].get_text().split(' ')[0] in end_date:
            temp_date = data_date[x].get_text().split(' ')[0]
            end_flag = True

        # 기사 끝날 날짜 체크(반복문 out)    
        if end_flag:
            if not(temp_date == data_date[x].get_text().split(' ')[0]):
                break       
        
        
        if not(start_flag):            
            # 기사 시작 날짜 체크
            if not(data_date[x].get_text().split(' ')[0] in start_date):            
                continue
            print("%d 페이지 수집중입니다.... ing" %page, end='')
            start_flag = True        
  
        
        # 기사 제목만 가져오기
        title.append(re.sub('[\'\n\r\t\xa0]', '', data_title[x].get_text()))
   
        # 기사 날짜만 가져오기
        date.append(data_date[x].get_text().split(' ')[0])
        
        if choose == '2':
            # 뉴스본문 url만 가져오기
            news_urls = 'http://www.hani.co.kr' + data_title[x].find('a')['href']
        
            # 뉴스본문 가져오기
            html = urlopen(news_urls)
            news_data = BeautifulSoup(html, "lxml")
            news_temp = news_data.find_all('div', 'text')
            news.append(re.sub('[\'\n\r\t\xa0]', '', news_temp[0].get_text()))
        
        

    # 데이터 수집 시작 체크
    if not(start_flag):
        continue
        
    # 데이터수집 체크
    print("     [수집완료]")       
          
    
    # 기사 끝날 날짜 체크(반복문 out)  
    if end_flag:
        if not(temp_date == data_date[x].get_text().split(' ')[0]):
            break 

        
end = datetime.now()
print("=" * 50)
print("데이터 수집을 완료했습니다.")
elapsed = end - start
print('총 계산 시간: ', end='');   print(elapsed)
print('총 수집 개수: %d개' %len(title))
```

```python
import pandas as pd
import re
```

```python
poli = pd.read_csv('./teample_1_result/hankyure_politics_commonword_(2018-01-01 - 2018-11-15).csv', sep=(','), encoding='utf-8')
eco = pd.read_csv('./teample_1_result/hankyure_economy_commonword_(2018-01-01 - 2018-11-15).csv', sep=(','), encoding ='utf-8')
soci = pd.read_csv('./teample_1_result/hankyure_society_commonword_(2018-01-01 - 2018-11-15).csv', sep=(','), encoding='utf-8')
# opi = pd.read_csv('./data/chosun_opinion_commonword.csv', sep=(','), encoding='utf-8')
```

#### 데이터프레임화

```python
poli_df = pd.DataFrame(poli)
eco_df = pd.DataFrame(eco)
soci_df = pd.DataFrame(soci)
# opi_df = pd.DataFrame(opi)
```

#### 불필요한 열 제거

```python
poli_df.drop('Unnamed: 0',axis=1, inplace=True)
eco_df.drop('Unnamed: 0',axis=1, inplace=True)
soci_df.drop('Unnamed: 0',axis=1, inplace=True)
# opi_df.drop('Unnamed: 0',axis=1, inplace=True)
```

#### 컬럼 명 변경

```python
poli_df.rename(columns={'단어명':'단어명', '빈도수':'정치빈도수'}, inplace=True)
eco_df.rename(columns={'단어명':'단어명', '빈도수':'경제빈도수'}, inplace=True)
soci_df.rename(columns={'단어명':'단어명', '빈도수':'사회빈도수'}, inplace=True)
# opi_df.rename(columns={'단어명':'단어명', '빈도수':'사설빈도수'}, inplace=True)
```

#### '단어명' 기준으로 병합

```python
from functools import reduce
import pandas as pd
dfs = [poli_df, eco_df, soci_df]
df_merge = reduce(lambda left, right: pd.merge(left, right, on='단어명'), dfs)
```

```python
map_1 = folium.Map(location=[45.372, -121.6972], tiles='Stamen Terrain', zoom_start=12)
folium.Marker([45.3288, -121.6625], popup='Mt. Hood Meadows').add_to(map_1)
folium.Marker([45.3311, -121.7113], popup='Timberline Lodge').add_to(map_1)
map_1
```

```python
map_2 = folium.Map(location=[45.372, -121.6972], tiles='Stamen Terrain', zoom_start=12)
folium.Marker([45.3288, -121.6625], popup='Mt. Hood Meadows', 
              icon=folium.Icon(icon='cloud')).add_to(map_2) # icon모양이 구름으로 바뀐다.
folium.Marker([45.3311, -121.7113], popup='Timberline Lodge', 
              icon=folium.Icon(icon='cloud')).add_to(map_2)
map_2
```

```python
map_3 = folium.Map(location=[45.372, -121.6972], zoom_start=12, tiles='Stamen Terrain')
folium.Marker([45.3288, -121.6625], popup='Mt. Hood Meadows', 
              icon=folium.Icon(icon='cloud')).add_to(map_3)
folium.Marker([45.3311, -121.7113], popup='Timberline Lodge', 
              icon=folium.Icon(color='green')).add_to(map_3)
folium.Marker([45.3300, -121.6823], popup='Some Other Location', 
              icon=folium.Icon(color='red',icon='info-sign')).add_to(map_3)
map_3
```

```python
map_4 = folium.Map(location=[45.5236, -122.6750], tiles='Stamen Toner', zoom_start=13)
folium.Marker([45.5244, -122.6699], popup='The Waterfront' ).add_to(map_4)
folium.CircleMarker([45.5215, -122.6261], radius=50, 
                    popup='Laurelhurst Park', color='#3186cc', fill_color='#3186cc', ).add_to(map_4)
map_4
```

```python
map_5 = folium.Map(location=[45.5236, -122.6750], zoom_start=13)
folium.RegularPolygonMarker([45.5012, -122.6655], popup='Ross Island Bridge', 
                            fill_color='#132b5e', number_of_sides=3, radius=10).add_to(map_5)
folium.RegularPolygonMarker([45.5132, -122.6708], popup='Hawthorne Bridge', 
                            fill_color='#45647d', number_of_sides=4, radius=10).add_to(map_5)
folium.RegularPolygonMarker([45.5275, -122.6692], popup='Steel Bridge', 
                            fill_color='#769d96', number_of_sides=6, radius=10).add_to(map_5)
folium.RegularPolygonMarker([45.5318, -122.6745], popup='Broadway Bridge', 
                            fill_color='#769d96', number_of_sides=8, radius=10).add_to(map_5)
map_5
```

# 서울시 구별 빈도수 시각화 
> 서울시 구별 무료와이파이존의 수를 지도에 나타낸다.

### 서울시 공공데이터 3
> -서울형 지도태깅 서울 Wi-Fi 정보
- Datasets 다운 : http://data.seoul.go.kr/dataList/datasetView.do?infId=OA-13591&srvType=S&serviceKind=1&currentPageNo=1
- 파일이름 : data/seoul_free_wifi_zone_2017.csv

```python
# Jupyter 실행시에 Warning 을 없애고 싶을때
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
import json
geo_path = 'data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
import pandas as pd
df = pd.read_csv('data/seoul_wifi_zone.csv')
df.head()
```

```python
df.columns = (['구', '유형', '위치명', 'lat', 'lon', '설치회사'])
df.head()
```

```python
freq_gu = df['구'].value_counts()
freq_gu
```

```python
map = folium.Map(location=[37.5102, 126.982], zoom_start=11)
map
```

```python
map = folium.Map(location=[37.5102, 126.982], zoom_start=11, tiles='Stamen Toner')
map
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map.choropleth(geo_data=geo_str,
              data=freq_gu,
              columns=['gu', 'counts'],
              fill_color='YlGnBu', #PuRd, YlGnBu
              key_on='feature.id')

map

# geo_str이라는 json 파일과 결합하는 과정을 추가한 코드이다.
# json  파일의 gu 에 해당하는 영역과 freq_gu의 counts와 columns로 연결.
# 그 크기만큼을 등급만큼 yellow부터 blue로 채워나가는 것을 덧씌우는 것.
# feature.id는 json 파일의 id 를 key값으로 연결하란 뜻인 듯.
```

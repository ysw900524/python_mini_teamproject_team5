### 워드클라우드 : USA Constitution 

```python
with open('wc_docu/constitution.txt') as fp:
    text = fp.read()
```

```python
len(text), text[:1000]
```

```python
# WordCloud 자체가 형태소 분석하여 빈도수 계산
wordcolud_1 = WordCloud().generate(text) 
type(wordcolud_1)
```

```python
wordcolud_1.words_
```

```python
save_img = 'img/wordcolud_constitution_1.png'

plt.figure(figsize=(12,12))
plt.imshow(wordcolud_1)
plt.axis('off') # 이미지 옆에 축 단위가 표시되는지 안되는지 설정 여부
plt.savefig(save_img)
```

```python
# WordCloud 자체가 형태소 분석하여 빈도수 계산
wordcolud_2 = WordCloud(width=800, height=600, background_color="purple")
wordcolud_2 = wordcolud_2.generate(text) 
type(wordcolud_2)
```

```python
save_img = 'img/wordcolud_constitution_2.png'

plt.figure(figsize=(24,16))
plt.imshow(wordcolud_2)
plt.axis('on')
plt.savefig(save_img)
```

```python
# WordCloud 자체가 형태소 분석하여 빈도수 계산
params = {
    "width"  : 800,
    "height" : 600, 
    "max_words" : 400,
    "colormap"  : "magma", 
    "background_color" :"orange"
}

wordcolud_3 = WordCloud(**params) #unpacking을 통해서 params를 뿌려준다.
wordcolud_3 = wordcolud_3.generate(text) 

save_img = 'img/wordcolud_constitution_3.png'
plt.figure(figsize=(24,16))
plt.imshow(wordcolud_3)
plt.axis('off')
plt.savefig(save_img)
```

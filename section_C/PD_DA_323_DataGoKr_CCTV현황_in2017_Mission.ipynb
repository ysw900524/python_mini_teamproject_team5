\# 경찰서 별 개수

```python
df_index = pd.Series(df['개소'][0:31].values, index=df['경찰서'][0:31].values)
df_index
```

```python
df_index.values
```

```python
for n in np.arange(len(df_index)):
    df_index.values[n] = int(df_index.values[n].replace(',',''))
```

```python
df_index.values
```

```python
plt.rcParams["figure.figsize"]=[12, 6]
df_index.plot(kind='barh', alpha=0.4)
```

\# cctv 개수

```python
df_index2 = pd.Series(df['대수'][0:31].values, index = df['경찰서'][0:31].values)
df_index2
```

```python
for n in np.arange(len(df_index2)):
    df_index2.values[n] = int(df_index2.values[n].replace(',',''))
```

```python
df_index2
```

```python
plt.rcParams["figure.figsize"]=[12, 6]
# plt.barh(df_index, width=0.7)
# plt.barh(df_index2, width=0.7)
df_index2.plot(kind='barh', alpha=0.6)
df_index.plot(kind='barh')

plt.show()
```

```python
plt.rcParams["figure.figsize"]=[12, 12]
df_index2.plot(kind='pie', title='CCTV 설치대수')
```

```python
# np.diff
Signature: np.diff(a, n=1, axis=-1)
Docstring:
Calculate the n-th discrete difference along given axis.

The first difference is given by ``out[n] = a[n+1] - a[n]`` along
the given axis, higher differences are calculated by using `diff`
recursively.


### 
diff 함수는 미분해주는 함수라고 알고 있으면 된다.
```

```python
dy = np.diff(y)     # numpy의 차분 함수 사용하기
                    # diff 함수는 미분해주는 함수라고 알고 있으면 된다.
dy[:10]
```

```python
dy = np.insert(dy, 0, 0)/0.01   # 차분의 특성으로 제일 앞에 의미없는 값 입력해두기
dy[:10]
```

```python
# np.insert

Signature: np.insert(arr, obj, values, axis=None)
Docstring:
Insert values along the given axis before the given indices.
```

```python
a = np.array([[1, 1], [2, 2], [3, 3]])
a
```

```python
np.insert(a, 2, 5)
```

```python
np.insert(a, 2, 5, axis=1)
# axis 값에 따라서 입력되는 방식이 달라짐을 알아야 한다.
```

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html?highlight=pie#matplotlib.pyplot.pie
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes  = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle. # equal 넣으면 위에서 figsize=(6,6) 넣은 것과 같은 효과

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 12), subplot_kw=dict(aspect="equal"))

recipe = ["375 g 밀가루",
          "75 g  설탕",
          "250 g 버터",
          "300 g 딸기"]

data = [float(x.split()[0]) for x in recipe]   # float 요소인 g앞의 수만 뽑는 코드(split으로 나누고 가장 먼저 나오는 [0]번째 요소 뽑기)
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="주요성분",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=14, weight="bold")

ax.set_title("빵의 주요 성분")

plt.show()
```

### annotation 적용 

```python
t = np.arange(0,5,0.01)
y = np.cos(2*np.pi*t)

plt.figure(figsize=(10,4))
plt.plot(t,y)
plt.annotate('local max', xy=(1,1), xytext=(1.5,1.2),
            arrowprops=dict(facecolor='black',shrink=0.05)) # 'local max'라는 글자를 가진 화살표를 만들 건데, xy 지점을 목표로 하며, 글자의 위치는 xytext 위치에 놓는다. 화살표 색깔은 검은색, 기울기는(shrink) 0.05다.
plt.ylim(-1.1,1.4)
plt.show()
```

### subplot 적용 

```python

# plt.subplot
Signature: plt.subplot(*args, **kwargs)
Docstring:
Return a subplot axes positioned by the given grid definition.

Typical call signature::

  subplot(nrows, ncols, plot_number)

Where *nrows* and *ncols* are used to notionally split the figure
into ``nrows * ncols`` sub-axes, and *plot_number* is used to identify
the particular subplot that this function is to create within the notional
grid. *plot_number* starts at 1, increments across rows first and has a
maximum of ``nrows * ncols``.
```

```python
plt.figure(figsize=(16,8))
plt.subplot(221)
plt.subplot(222)
plt.subplot(212)

plt.show()
```

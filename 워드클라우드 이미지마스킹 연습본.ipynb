```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('wc_docu/alice.txt') as fp:
    text = fp.read()
```

```python
len(text), text[:1000]
```

```python
# 사진 데이터를 pil 패키지의 Image를 활용해서 0~255 데이터로 읽어왔다.
starry_night = np.array(Image.open('images/starry_night.png'))
```

```python
starry_night
```

```python
# plt.imshow()는 수치 데이터 -> image데이터로 변환해줌
plt.imshow(starry_night, cmap=plt.cm.gray)
plt.axis('off')
```

```python
# STOPWORDS : 많은 빈도로 나오는 관사, 전치사 모아놓은 데이터
stopwords =set(STOPWORDS)
stopwords
```

```python
params = {
    "width"  : 800,
    "height" : 600, 
    "max_words" : 2000,
    "background_color" :"white",
    "mask"      : starry_night,
    "stopwords" : stopwords
}

# wordcolud_4 = WordCloud(background_color = 'white', max_words=2000, mask=alice_mask, stopwords=stopwords)
wordcolud_7 = WordCloud(**params)
wordcolud_7 = wordcolud_7.generate(text)
wordcolud_7.words_
```

```python
save_img = 'img/wordcloud_starry_night_1.png'

plt.figure(figsize=(24,24))
plt.imshow(wordcolud_7)
plt.axis('off')
plt.savefig(save_img)
```

```python
# with open('wc_docu/alice.txt') as fp:
#     text = fp.read()
# text = open('alice.txt').read()

alice_color = np.array(Image.open('images/alice_color.png'))

stopwords = set(STOPWORDS)

# 카운팅 하고싶지 않은 단어 추가
stopwords.add('said')       
stopwords.add('said Alice')
```

```python
params = {
    "width"     : 600,
    "height"    : 600, 
    "max_words" : 2000,
    "max_font_size" : 40, 
    "random_state"  : 42,
    "background_color" :"white",
    "mask"      : starry_night,
    "stopwords" : stopwords
}

wordcolud_8 = WordCloud(**params)
wordcolud_8 = wordcolud_8.generate(text)
# wordcolud_5.words_
```

```python
plt.figure(figsize=(1,1))
plt.imshow(starry_night)
plt.axis('off')
```

```python
save_img = 'img/wordcolud_starry_night_2.png'

plt.figure(figsize=(12, 12))
plt.imshow(wordcolud_8)
plt.axis('off')
plt.savefig(save_img)
```

```python
from wordcloud import ImageColorGenerator
```

```python
starry_night = np.array(Image.open('images/starry_night.png'))
image_colors = ImageColorGenerator(starry_night)
```

```python
save_img = 'img/wordcolud_starry_night_3.png'

plt.figure(figsize=(12,12))
# wordcloud_5.recolr(ImageColorGenerator(numpy이미지 파일))
plt.imshow(wordcolud_8.recolor(color_func=image_colors))
plt.axis('off')
plt.savefig(save_img)
```

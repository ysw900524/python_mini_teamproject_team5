```python
title = []
date = []
```

```python
def article_crawrling(x):
    for i in range(int(x)):
#         title = []
#         date = []
        
        url = 'http://www.hani.co.kr/arti/society/list{}.html'.format(i+1)
#         print(url)
        html = urlopen(url)
        # url = 'http://www.hani.co.kr/arti/politics/list{}.html'
        # html = urlopen(url.format())
        soup =BeautifulSoup(html, 'lxml')
#         print(soup)
        tmp = soup.find_all('div', 'article-area')
    
        for j in range(len(tmp)):
            tmp_title = tmp[j].find('h4', 'article-title')
            title.append(tmp_title.get_text().strip())
#             print(title)
            
            tmp_date = tmp[j].find('span', 'date')
#             if tmp_date.get_text().split(' ')[0] == '2018-11-08' :
#                 break
            date.append(tmp_date.get_text().split(' ')[0])            
        if date[-1] == '2018-11-08' :
            break
        
        print('{}번째 페이지를 크롤링 중입니다.'.format(i+1))
```

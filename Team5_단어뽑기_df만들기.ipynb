# 경제 리스트 데이터프레임 만들기

```python
word = pd.read_csv('./teample_1/hankyure_economy_uft.csv', sep=',', encoding='utf-8')
# word
```

```python
xyz = list(word['제목'])
```

```python
words = str()

for i in range(len(xyz)):
    words += xyz[i]

# words
```

# 트위터

```python
# from konlpy.tag import Hannanum
# from konlpy.tag import Kkma
# from konlpy.tag import Komoran
# from konlpy.tag import Mecab
from konlpy.tag import Twitter
```

```python
twitter = Twitter()
nouns3 = twitter.nouns(words)
count3 = Counter(nouns3)
```

```python
common_word_economy = count3.most_common()
common_word_economy
```

```python
# type(common_word_economy)
```

```python
# common_word_economy[0][0]
# common_word_economy[0][1]
```

```python
wordname = []
wordfreq = []

for i in range(len(common_word_economy)):
    wordname.append(common_word_economy[i][0])
    wordfreq.append(common_word_economy[i][1])
    
    ecolist = pd.DataFrame({'단어명' : wordname, '빈도수':wordfreq})
    
```

```python
ecolist.to_csv('./teample_1_commonword_result/economy_commonword_df.csv', sep=',', encoding='utf-8')
```

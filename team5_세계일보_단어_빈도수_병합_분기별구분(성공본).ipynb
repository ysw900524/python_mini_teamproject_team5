#### 데이터프레임화

```python
tmp1_df = pd.DataFrame(tmp1)
tmp2_df = pd.DataFrame(tmp2)
tmp3_df = pd.DataFrame(tmp3)
tmp4_df = pd.DataFrame(tmp4)
```

#### 불필요한 열 제거

```python
tmp1_df.drop('Unnamed: 0',axis=1, inplace=True)
tmp2_df.drop('Unnamed: 0',axis=1, inplace=True)
tmp3_df.drop('Unnamed: 0',axis=1, inplace=True)
tmp4_df.drop('Unnamed: 0',axis=1, inplace=True)
```

#### 컬럼 명 변경

```python
# # 경제
# tmp1_df.rename(columns={'단어명':'단어명', '빈도수':'경제1분기빈도수'}, inplace=True)
# tmp2_df.rename(columns={'단어명':'단어명', '빈도수':'경제2분기빈도수'}, inplace=True)
# tmp3_df.rename(columns={'단어명':'단어명', '빈도수':'경제3분기빈도수'}, inplace=True)
# tmp4_df.rename(columns={'단어명':'단어명', '빈도수':'경제4분기빈도수'}, inplace=True)

# # 사회
# tmp1_df.rename(columns={'단어명':'단어명', '빈도수':'사회1분기빈도수'}, inplace=True)
# tmp2_df.rename(columns={'단어명':'단어명', '빈도수':'사회2분기빈도수'}, inplace=True)
# tmp3_df.rename(columns={'단어명':'단어명', '빈도수':'사회3분기빈도수'}, inplace=True)
# tmp4_df.rename(columns={'단어명':'단어명', '빈도수':'사회4분기빈도수'}, inplace=True)

# 정치
tmp1_df.rename(columns={'단어명':'단어명', '빈도수':'정치1분기빈도수'}, inplace=True)
tmp2_df.rename(columns={'단어명':'단어명', '빈도수':'정치2분기빈도수'}, inplace=True)
tmp3_df.rename(columns={'단어명':'단어명', '빈도수':'정치3분기빈도수'}, inplace=True)
tmp4_df.rename(columns={'단어명':'단어명', '빈도수':'정치4분기빈도수'}, inplace=True)

```

#### '단어명' 기준으로 병합

```python
from functools import reduce
import pandas as pd
dfs = [tmp1_df, tmp2_df, tmp3_df, tmp4_df]
word_freq = reduce(lambda left, right: pd.merge(left, right, on='단어명'), dfs)
```

```python
word_freq = word_freq.sort_values('총빈도수', ascending=False)
```
